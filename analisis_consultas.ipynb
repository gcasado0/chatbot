```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_excel('datos/encontraste_lo_que_buscabas-gemini-2.0-flash-20250416_103831.xlsx')
```

```python
# eliminar registros que no tengan sentimeinto
data = data[data['Sentimiento'].notna()]
```

```python
print(data.head(5))
print(data.info())
```

```python
# Calcular cantidad y proporción por categoría en la columna 'Sentimiento'
resumen_df = (
    data['Sentimiento']
    .value_counts(normalize=False)
    .rename_axis('Sentimiento')
    .reset_index(name='Cantidad')
)
resumen_df['Proporcion'] = resumen_df['Cantidad'] *100/ resumen_df['Cantidad'].sum()

# Visualización
sns.set(style="whitegrid")
plt.figure(figsize=(8, 4))
sns.barplot(x='Sentimiento', y='Proporcion', data=resumen_df)
plt.title('Proporción de Sentimientos')
plt.xlabel('Sentimiento')
plt.ylabel('Proporción')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Calcular cantidad y proporción por categoría en la columna 'Enojo'
resumen_df = (
    data['Enojo']
    .value_counts(normalize=False)
    .rename_axis('Enojo')
    .reset_index(name='Cantidad')
)
resumen_df['Proporcion'] = resumen_df['Cantidad'] *100/ resumen_df['Cantidad'].sum()
# Visualización
plt.figure(figsize=(8, 4))
sns.barplot(x='Enojo', y='Proporcion', data=resumen_df)
plt.title('Proporción de Enojo')
plt.xlabel('Enojo')
plt.ylabel('Proporción')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show();

```

```python
# Calcular cantidad y proporción por categoría en la columna 'Principal' ordenando por cantidad
resumen_df = (
    data['Principal']
    .value_counts(normalize=False)
    .rename_axis('Principal')
    .reset_index(name='Cantidad')
)
resumen_df['Proporcion'] = resumen_df['Cantidad'] *100/ resumen_df['Cantidad'].sum()
# Visualización
plt.figure(figsize=(8, 4))
sns.barplot(x='Principal', y='Proporcion', data=resumen_df)
plt.title('Distribucón de Categoria Principal')
plt.xlabel('Principal')
plt.ylabel('Proporción')
plt.xticks(rotation=45)
# Hacer que si el valor es igual a "Indefinido" se muestre en rojo tenue y con el fondo de la barra en rojo tenue
plt.bar(resumen_df['Principal'], resumen_df['Proporcion'], color=['red' if x == 'Indefinido' else 'blue' for x in resumen_df['Principal']])
# Añadir etiquetas a las barras
for index, value in enumerate(resumen_df['Proporcion']):
    if resumen_df['Principal'][index] == 'Indefinido':
        plt.text(index, value, f'{value:.2f}%', color='red', ha='center')
    else:
        plt.text(index, value, f'{value:.2f}%', ha='center')
plt.tight_layout()
plt.show();
```

```python
# Calcular cantidad y proporción por categoría en la columna 'Secundaria' ordenando por cantidad y filtrado por Principal='Reclamos'
resumen_df = (
    data[data['Principal'] == 'Reclamos']['Secundaria']
    .value_counts(normalize=False)
    .rename_axis('Secundaria')
    .reset_index(name='Cantidad')
)
resumen_df['Proporcion'] = resumen_df['Cantidad'] *100/ resumen_df['Cantidad'].sum()
# Visualización
plt.figure(figsize=(8, 4))
sns.barplot(x='Secundaria', y='Proporcion', data=resumen_df)
plt.title('Proporción de Categoria Secundaria de Reclamos')   
plt.xlabel('Secundaria')
plt.ylabel('Proporción')
plt.xticks(rotation=45)
# Hacer que si el valor es igual a "Indefinido" se muestre en rojo
for index, value in enumerate(resumen_df['Proporcion']):
    if resumen_df['Secundaria'][index] == 'Indefinido':
        plt.text(index, value, f'{value:.2f}%', color='red', ha='center')
    else:
        plt.text(index, value, f'{value:.2f}%', ha='center')
plt.tight_layout()
plt.show();

```

```python
# Calcular cantidad y proporción por categoría en la columna 'Secundaria' ordenando por cantidad y filtrado por Principal='Reclamos'
resumen_df = (
    data[data['Principal'] == 'Multas']['Secundaria']
    .value_counts(normalize=False)
    .rename_axis('Secundaria')
    .reset_index(name='Cantidad')
)
resumen_df['Proporcion'] = resumen_df['Cantidad'] *100/ resumen_df['Cantidad'].sum()
# Visualización
plt.figure(figsize=(8, 4))
sns.barplot(x='Secundaria', y='Proporcion', data=resumen_df)
plt.title('Proporción de Categoria Secundaria de Multas')   
plt.xlabel('Secundaria')
plt.ylabel('Proporción')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show();
```
